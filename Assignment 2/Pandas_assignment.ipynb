```python
# Import pandas
import pandas as pd
```

```python
# Create a series of fruits(string) of length 5 and view it
# Create a series of prices(int) of length 5 and view it
fruits = pd.Series(["Apple", "Orange", "Grapes", "Strawberry", "Mangoes"])

prices = pd.Series([100, 50 ,70, 40, 60])

print(fruits)
print(prices)
```

```python
# Create a fruit_prices dataframe out of the 2 series you just made and view it
data = {"Fruits":fruits, "Prices":prices}
fruit_prices = pd.DataFrame(data)
fruit_prices
```

```python
# Find the mean, sum, standard deviation, variance of the prices
print("Mean :")
print(fruit_prices.mean())

print("\nSum :")
print(fruit_prices.sum())

print("\nStandard Devition :")
print(fruit_prices.std())

print("\nVariance :")
print(fruit_prices.var())
```
